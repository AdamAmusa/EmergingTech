```python
from collections import defaultdict

with open('englishworks/Dracula.txt', 'r') as file:
    english = file.read()

english = english.upper()



keep = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ .'

cleaned = ''.join(c for c in english if c in keep)

```

## Creating a TriGram Model

The following was adapted from a response from ChatGPT  
https://chatgpt.com/

```python
def createTrigrams(text):
    trigram_model = defaultdict(int) #initialise empty dictionary to store trigram model

    for i in range(len(text) -2):
        trigram = text[i:i+3] #Splices text into three characters

        trigram_model[trigram] += 1  #counts after each increment the number of time the three characters appear
    
    return dict(trigram_model)  #returns dictionary of trigram model


createTrigrams(cleaned)
        
```
