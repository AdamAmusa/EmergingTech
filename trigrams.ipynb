```python
from collections import defaultdict

dracula = "englishworks/dracula.txt"
christmascarol = "englishworks/christmascarol.txt"
frankenstein = "englishworks/frankenstein.txt"
peterpan = "englishworks/peterpan.txt"
sherlockholmes = "englishworks/sherlockholmes.txt"

with open(frankenstein, 'r') as file:
    english = file.read()

english = english.upper()

keep = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ .'

cleaned = ''.join(c for c in english if c in keep)

```

## Creating a TriGram Model

The following was adapted from a response from ChatGPT  
https://chatgpt.com/

```python
def createTrigrams(text):
    #initialise empty dictionary to store trigram model
    trigram_model = defaultdict(int) 

    #loop over the length of the text while leaving the last two characters
    for i in range(len(text) -2): 
        #Splices text into three characters
        trigram = text[i:i+3]
        #counts after each increment the number of time the three characters appear as well as initialising keys
        trigram_model[trigram] += 1

    #returns dictionary of trigram model
    return dict(trigram_model)  


count = createTrigrams(cleaned)
```
