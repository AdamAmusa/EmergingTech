```python
from collections import defaultdict
import random

#Eglish Text to choose from
dracula = "englishworks/dracula.txt"
christmascarol = "englishworks/christmascarol.txt"
frankenstein = "englishworks/frankenstein.txt"
peterpan = "englishworks/peterpan.txt"
sherlockholmes = "englishworks/sherlockholmes.txt"

#opens file and stores the read into a variable
with open(frankenstein, 'r') as file:
    english = file.read()

#convert contents to Upper case
english = english.upper()

#Character we want to keep
keep = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ .'

#Loop through the file, If the characters are included in keep join them into the variable
cleaned = ''.join(c for c in english if c in keep)

```

## Creating a TriGram Model

The following was adapted from a response from ChatGPT  
https://chatgpt.com/

```python
def createTrigrams(text):
    #initialise empty dictionary to store trigram model
    trigram_model = defaultdict(int) 

    #loop over the length of the text while leaving the last two characters
    for i in range(len(text) -2): 
        #Splices text into three characters
        trigram = text[i:i+3]
        #counts after each increment the number of time the three characters appear as well as initialising keys
        trigram_model[trigram] += 1

    #returns dictionary of trigram model
    return dict(trigram_model)  


count = createTrigrams(cleaned)
```

```python
#Make weights equal to the value so we can use it to weigh the possibility of the characters
#Make chars equal to the keys so we can output the characters
chars, weights = count.keys(), count.values() 
```

The following code was adapted from the [lab notes](https://github.com/ianmcloughlin/2425_emerging_technologies/blob/main/02_language_models.ipynb). Which implemented Python Random choices() from https://docs.python.org/3/library/random.html

```python

#Stores a sequence of characters from the list of "chars" using weights to weigh the frequency of characters
L = random.choices(list(chars), weights=weights, k=10000)
```

```python
#Join the output into a string
s = ''.join(L)
s
```

Code improved by [ChatGpt](https://chatgpt.com/c/671247b2-6860-800c-905a-c3ee50d4e868) by converting list of words into a [set](https://docs.python.org/2/library/sets.html) to look up words in constant time and removing nested loop and replacing it with a [generator expression](https://wiki.python.org/moin/Generators) to calculate the sum

```python
with open("words.txt", "r") as file:
    words = file.read()

def percentageofWords(words, s):
    #assign words from the file into lists and seperate by spaces
    set_of_words = set(words.split())
    #assign words from the generation into lists and seperate by spaces
    generation = s.split()
    
    #loop through each word from the list of words and calculate sum if the word generation is an english word
    sum_of_percentage = sum(1 for g in generation if g in set_of_words)

    #return the percentage of real words from the generation and round it to two decimal places        
    return round(((sum_of_percentage/len(generation)) * 100),3)



percentageofWords(words, s)
```
